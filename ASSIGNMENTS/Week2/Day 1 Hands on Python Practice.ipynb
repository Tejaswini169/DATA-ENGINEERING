**1. If Statement**

```python
fruit = "apple"
if fruit == "apple":
    print("This is an apple!")
```

**2. If-Else Statement**

```python
fruit = "orange"
if fruit == "apple":
    print("This is an apple!")
else:
    print("This is not an apple!")
```

**3. If-Elif-Else Statement**

```python
fruit = "banana"
if fruit == "apple":
    print("This is an apple!")
elif fruit == "banana":
    print("This is a banana!")
else:
    print("This is some other fruit!")
```

**4. Control Structure**

Control structures include if statements, loops, etc., which control the flow of the program. Here's an example using an if statement within a loop.

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    if fruit == "banana":
        print("Found a banana!")
```

**5. For Loop**

```python
vegetables = ["carrot", "potato", "broccoli"]
for veggie in vegetables:
    print("I have a", veggie)
```

**6. While Loop**

```python
count = 3
while count > 0:
    print("Eating a snack")
    count -= 1
```

**7. Nested Loop**

```python
fruits = ["apple", "banana"]
toppings = ["chocolate", "whipped cream"]
for fruit in fruits:
    for topping in toppings:
        print(f"{fruit} with {topping}")
```

**8. Break, Continue & Pass**

```python
fruits = ["apple", "orange", "banana", "grape"]
for fruit in fruits:
    if fruit == "orange":
        continue  # Skip "orange"
    elif fruit == "banana":
        break    # Stop the loop at "banana"
    else:
        pass    # Do nothing for other fruits
    print(fruit)
```

**13. Introduction to Set & Set Methods**

```python
# Unique set of food items
food_set = {"apple", "banana", "apple", "grape"}
food_set.add("orange")
food_set.discard("banana")
print(food_set)
```

**14. Introduction to Map & Map Methods**

```python
# Mapping food items to uppercase
foods = ["apple", "banana", "cherry"]
uppercased_foods = list(map(str.upper, foods))
print(uppercased_foods)
```
