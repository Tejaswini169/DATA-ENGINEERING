```python
#left outer join & pandas left join
import pandas as pd
val1 = {'id': [1, 2, 10, 12], 
        'val1': ['a', 'b', 'c', 'd']}

df1 = pd.DataFrame (val1)

val2 ={'id': [1, 2, 9, 8],
       'val1': ['p', 'q', 'r', 's']}

df2= pd.DataFrame(val2)

df= pd.merge(df1, df2, on='id', how = 'left')

print(df)

print(df1)
print(df2)
```

```python
#right outer join & pandas left join
import pandas as pd
val1 = {'id': [1, 2, 10, 12], 
        'val1': ['a', 'b', 'c', 'd']}

df1 = pd.DataFrame (val1)

val2 ={'id': [1, 2, 9, 8],
       'val1': ['p', 'q', 'r', 's']}

df2= pd.DataFrame(val2)

df= pd.merge(df1, df2, on='id', how = 'right')

print(df)

print(df1)
print(df2)
```

```python
#outer join 
import pandas as pd
val1 = {'id': [1, 2, 10, 12], 
        'val1': ['a', 'b', 'c', 'd']}

df1 = pd.DataFrame (val1)

val2 ={'id': [1, 2, 9, 8],
       'val1': ['p', 'q', 'r', 's']}

df2= pd.DataFrame(val2)

df= pd.merge(df1, df2, on='id', how = 'outer')

print(df)

print(df1)
print(df2)
```

```python
#inner join 
import pandas as pd
val1 = {'id': [1, 2, 10, 12], 
        'val1': ['a', 'b', 'c', 'd']}

df1 = pd.DataFrame (val1)

val2 ={'id': [1, 2, 9, 8],
       'val1': ['p', 'q', 'r', 's']}

df2= pd.DataFrame(val2)

df= pd.merge(df1, df2, on='id', how = 'inner')

print(df)

print(df1)
print(df2)
```

```python
#left outer join & pandas left join
import pandas as pd
val1 = {'id': [1, 2, 10, 12], 
        'val1': ['a', 'b', 'c', 'd']}

df1 = pd.DataFrame (val1)

val2 ={'id': [1, 2, 9, 8],
       'val1': ['p', 'q', 'r', 's']}

df2= pd.DataFrame(val2)

df= pd.merge(df1, df2, left_index=True, right_index=True)

print(df)

print(df1)
print(df2)
```
